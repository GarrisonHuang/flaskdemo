```python
from bs4 import BeautifulSoup
import requests

url = "https://histock.tw/國際股市"
headers = {
    "User-Agent": "Mozilla/5.0"
}
resp = requests.get(url, headers=headers)
soup = BeautifulSoup(resp.text, "html.parser")

# 找到 <h3>今年以來</h3>
h3_tag = soup.find("h3", string="今年以來")
if h3_tag:
    # 找到這個 h3 的下一個 div（應該是 <div class="index-list">）
    index_div = h3_tag.find_next_sibling("div", class_="index-list")
    if index_div:
        rows = index_div.find_all("tr")[1:]  # 第一個是表頭，跳過
        for row in rows:
            cols = row.find_all("td")
            name = cols[0].text.strip()
            percent = cols[1].text.strip()
            print(name, percent)
    else:
        print("找不到 index-list div")
else:
    print("找不到 h3: 今年以來")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time

# 啟動 Selenium（自動下載 ChromeDriver）
options = webdriver.ChromeOptions()
options.add_argument("--headless")  # 無頭模式，不會打開瀏覽器畫面
options.add_argument("--disable-gpu")
options.add_argument("--no-sandbox")

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
driver.get("https://histock.tw/國際股市")

# 等待 JavaScript 把資料載入
time.sleep(3)

# 取得完整渲染後的 HTML
html = driver.page_source
driver.quit()

# 用 BeautifulSoup 解析
soup = BeautifulSoup(html, "html.parser")

# 找 <h3>今年以來</h3>
h3_tag = soup.find("h3", string="今年以來")
if h3_tag:
    index_div = h3_tag.find_next_sibling("div", class_="index-list")
    if index_div:
        rows = index_div.find_all("tr")[1:]  # 跳過表頭
        for row in rows:
            cols = row.find_all("td")
            name = cols[0].text.strip()
            percent = cols[1].text.strip()
            print(name, percent)
    else:
        print("找不到 index-list div")
else:
    print("找不到 h3: 今年以來")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup

options = webdriver.ChromeOptions()
options.add_argument("--headless")
options.add_argument("--disable-gpu")
options.add_argument("--no-sandbox")

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
driver.get("https://histock.tw/國際股市")

datas = []

try:
    # 等待資料載入
    WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, "//h3[text()='今年以來']/following-sibling::div[@class='index-list']"))
    )

    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")

    h3_tag = soup.find("h3", string="今年以來")
    if h3_tag:
        index_div = h3_tag.find_next_sibling("div", class_="index-list")
        if index_div:
            trs = index_div.find_all("tr")

            for tr in trs:
                data = []

                # 先抓 <th>（表頭）
                ths = tr.find_all("th")
                if ths:
                    for th in ths:
                        data.append(th.text.strip())

                # 再抓 <td>（數值）
                tds = tr.find_all("td")
                if tds:
                    for td in tds:
                        data.append(td.text.strip())

                if data:  # 如果有資料才加入 datas
                    datas.append(data)

    # ✅ 印出 datas 結構
    for row in datas:
        print(row)

finally:
    driver.quit()
```

```python
def convert_value(value):
    try:
        return eval(value)
    except:
        return None
```

```python
df["pm25"] = df["pm25"].apply(convert_value)
# 移除有None 的數據
df = df.dropna()
df
```
